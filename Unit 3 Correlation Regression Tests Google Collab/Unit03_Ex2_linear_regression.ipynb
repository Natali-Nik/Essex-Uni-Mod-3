```python
# prompt: perform an EDA

import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import numpy as np

# Create a Pandas DataFrame for easier data manipulation
data = {'x': [5,7,8,7,2,17,2,9,4,11,12,9,6],
        'y': [99,86,87,88,111,86,103,87,94,78,77,85,86]}
df = pd.DataFrame(data)

# Summary statistics
print(df.describe())

# Correlation analysis
correlation = df['x'].corr(df['y'])
print(f"Correlation between x and y: {correlation}")


# Linear Regression
slope, intercept, r_value, p_value, std_err = stats.linregress(df['x'], df['y'])
print(f"Slope: {slope}, Intercept: {intercept}, R-value: {r_value}, P-value: {p_value}, Standard error: {std_err}")


# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(df['x'], df['y'], label='Data Points')
plt.plot(df['x'], intercept + slope * df['x'], 'r', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Residual analysis
residuals = df['y'] - (intercept + slope * df['x'])
plt.figure(figsize=(8, 4))
plt.scatter(df['x'], residuals)
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at y=0
plt.xlabel('x')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()


# Histogram of residuals
plt.figure(figsize=(8, 4))
plt.hist(residuals, bins=10) # Adjust bins as needed
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()


# QQ plot of residuals
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Q-Q Plot of Residuals")
plt.show()
```

```python
# prompt: adjust data set so that x and y have similar maximum values

# Find the maximum values of x and y
max_x = df['x'].max()
max_y = df['y'].max()

# Determine the scaling factor
scaling_factor = max_x / max_y

# Scale the 'y' values
df['y_scaled'] = df['y'] * scaling_factor

# Now x and y_scaled have similar maximum values
print(df)

# Recalculate statistics and plots using 'y_scaled'
slope, intercept, r_value, p_value, std_err = stats.linregress(df['x'], df['y_scaled'])

plt.figure(figsize=(10, 6))
plt.scatter(df['x'], df['y_scaled'], label='Data Points')
plt.plot(df['x'], intercept + slope * df['x'], 'r', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y_scaled')
plt.title('Linear Regression (Scaled y)')
plt.legend()
plt.grid(True)
plt.show()
```
